```python
# imports and nonsense that you only want to run once
import os
import glob
# import transformers
from transformers import BertTokenizer, BertForMaskedLM
```

```python
# cell to define all function
def LoadTextChars(fpath):
    '''Nothing fancy text reader. we want to load text
    so that it is already structured just get character sequences
    hence f.read() and not f.readlines()
    '''
    with open(fpath,'r') as f:
        text = f.read()
    return text

def actions():
  # public domain books
  bookdir = '/content/drive/My Drive/Books/'
  books = sorted(glob.glob(os.path.join(bookdir, '*.txt')))
  text_0 = LoadTextChars(books[0])

  snippet = text_0[1000:1500]
  print('book name: ', books[0].split('/')[-1])
  print(snippet)
  tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
  tokenized_text = tokenizer(snippet)
  # You get a dictionary with input ids, token_type ids, and attention mask interesting
  print(tokenized_text)

# run the actions function to call everything you want to call
actions()
```
